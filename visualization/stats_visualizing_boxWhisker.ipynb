# SECTION: Visualizing data
### VIDEO: Box-and-whisker plots


```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
## create data for the bar plot

# data sizes
m = 30 # rows
n =  6 # columns

# generate data
data = np.zeros((m,n))

for i in range(n):
    data[:,i] = 30*np.random.randn(m) * (2*i/(n-1)-1)**2 + (i+1)**2
```

```python
# now for the boxplot

plt.boxplot(data)
plt.show()
```

```python
# now with seaborn
sns.boxplot(data=data,orient='v')
plt.show()
```

```python
# or as a pandas data frame
df = pd.DataFrame(data,columns=['zero','one','two','three','four','five'])
sns.boxplot(data=df,orient='h')
plt.show()
```

```python
# Generate some random data
np.random.seed(42)
data = np.random.normal(100, 20, 200)

# Create the box-and-whiskers plot
fig, ax = plt.subplots()

# Create the boxplot
box = ax.boxplot(data, patch_artist=True, notch=True, flierprops=dict(markerfacecolor='r', marker='o'))

# Annotate the boxplot
for i in range(len(data)):
    if data[i] < np.percentile(data, 25) - 1.5 * (np.percentile(data, 75) - np.percentile(data, 25)) or \
       data[i] > np.percentile(data, 75) + 1.5 * (np.percentile(data, 75) - np.percentile(data, 25)):
        ax.annotate(f'Outlier: {data[i]:.1f}', xy=(1, data[i]), xytext=(1.1, data[i]),
                    arrowprops=dict(facecolor='red', shrink=0.05))

# Customize colors for the box
for element in ['boxes', 'whiskers', 'fliers', 'medians', 'caps']:
    plt.setp(box[element], color='black')

# Set colors for different parts of the boxplot
colors = ['skyblue']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

# Annotate the median, quartiles, min, and max
median = np.median(data)
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
whisker_min = np.min(data[data >= q1 - 1.5 * (q3 - q1)])
whisker_max = np.max(data[data <= q3 + 1.5 * (q3 - q1)])

ax.annotate(f'Median: {median:.1f}', xy=(1, median), xytext=(1.05, median + 5),
            arrowprops=dict(facecolor='blue', shrink=0.05))

ax.annotate(f'Q1: {q1:.1f}', xy=(1, q1), xytext=(1.05, q1 - 10),
            arrowprops=dict(facecolor='green', shrink=0.05))

ax.annotate(f'Q3: {q3:.1f}', xy=(1, q3), xytext=(1.05, q3 + 5),
            arrowprops=dict(facecolor='green', shrink=0.05))

ax.annotate(f'Min: {whisker_min:.1f}', xy=(1, whisker_min), xytext=(1.05, whisker_min +5),
            arrowprops=dict(facecolor='purple', shrink=0.05))

ax.annotate(f'Max: {whisker_max:.1f}', xy=(1, whisker_max), xytext=(1.05, whisker_max -10),
            arrowprops=dict(facecolor='purple', shrink=0.05))

# Add labels
ax.set_title('Box-and-Whiskers Plot with Annotations')
ax.set_ylabel('Value')

# Show the plot
plt.show()
```
