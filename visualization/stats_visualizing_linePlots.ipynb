# SECTION: Visualizing data
# Line plots


```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
## create data for the plot

# number of data points
n = 1000

# generate log-normal distribution
data1 = np.exp( np.random.randn(n)/2 )
data2 = np.exp( np.random.randn(n)/10 )
data3 = np.exp( np.random.randn(n)/2 + 1 )
```

```python
## plots of their histograms

# number of histogram bins
k = 20

plt.hist(data1,bins=k)
plt.hist(data2,bins=k)
plt.hist(data3,bins=k)

plt.show()
```

```python
# histogram discretization for the datasets
y1,x1 = np.histogram(data1,bins=k)
xx1 = (x1[0:-1] + x1[1:]) / 2
y1 = y1 / sum(y1) # convert to probability

y2,x2 = np.histogram(data2,bins=k)
xx2 = (x2[0:-1] + x2[1:]) / 2
y2 = y2 / sum(y2) # convert to probability

y3,x3 = np.histogram(data3,bins=k)
xx3 = (x3[0:-1] + x3[1:]) / 2
y3 = y3 / sum(y3) # convert to probability



# show the plots
plt.plot(xx1,y1,'s-',label='data1')
plt.plot(xx2,y2,'o-',label='data2')
plt.plot(xx3,y3,'^-',label='data3')

plt.legend()
plt.xlabel('Value')
plt.ylabel('Probability')
plt.show()
```

# Code challenge: log-scaled plots
---

```python
#generate data
data_lin = np.linspace(start = 0.001, stop = 10, num = 100)
data_log = np.exp(data_lin)

# linear scale
plt.plot(data_log[:40])
plt.plot(data_lin)
plt.show()

# log scale
plt.plot(data_log)
plt.plot(data_lin)
plt.yscale("log" )
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 6))
ax[0].plot(data_lin, label="$x=x$")
ax[0].plot(data_log[:40], label="$x=e^x$")
ax[0].set_ylabel('Linear-scale')
ax[0].legend()

ax[1].plot(data_lin, label="$x=x$")
ax[1].plot(data_log, label="$x=e^x$")
ax[1].set_ylabel('Log-scale')
ax[1].set_yscale('log')
ax[1].legend()
plt.show();
```
