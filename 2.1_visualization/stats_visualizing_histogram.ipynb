# SECTION: Visualizing data
# Histograms

```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
## create data for the histogram

# number of data points
n = 1000

# generate data - log-normal distribution
data = np.exp( np.random.randn(n)/2 )
```

```python
# show as a histogram

# number of histogram bins
k = 50

plt.hist(data,bins=k)
plt.show()
```

```python
# another option
y,x = np.histogram(data,bins=k)

# bin centers
xx = (x[1:]+x[:-1])/2

plt.plot(xx,y)
plt.show()
```

# Code challenge: Histogram proportion
---

```python
# number of data points
n = 1000
# generate data for histogram - log-normal distribution
data = np.exp( np.random.randn(n)/2 )
```

```python
# number of histogram bins
k = 40

# plot histogram
plt.hist(data, bins=k)
```

```python
counts, bin_edges = np.histogram(data, bins = k)

# proportions = counts / n
proportions = counts / np.sum(counts)

# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html
plt.hist(bin_edges[:-1], bin_edges, weights = proportions)
```

```python
bin_centers = (bin_edges[1:] + bin_edges[:-1])/2
plt.plot(bin_centers, counts)
plt.show()
```

```python
plt.plot(bin_centers, proportions)
plt.show()
```
