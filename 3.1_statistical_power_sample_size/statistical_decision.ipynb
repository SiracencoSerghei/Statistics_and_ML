```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Parameters for the distributions
mu_null = 0  # Mean under null hypothesis (H0)
mu_alt = 2   # Mean under alternative hypothesis (H1)
sigma = 1    # Standard deviation
alpha = 0.05  # Significance level (Type I error threshold)
x = np.linspace(-3, 5, 500)  # X-axis values

# PDF for null and alternative hypotheses
pdf_null = norm.pdf(x, mu_null, sigma)
pdf_alt = norm.pdf(x, mu_alt, sigma)

# Critical value for rejecting H0 (right-tailed test for simplicity)
z_critical = norm.ppf(1 - alpha, mu_null, sigma)

# Plot
plt.figure(figsize=(12, 6))

# Plot distributions
plt.plot(x, pdf_null, label='H0 (Null Hypothesis)', color='blue', linewidth=2)
plt.plot(x, pdf_alt, label='H1 (Alternative Hypothesis)', color='red', linewidth=2)

# Fill areas for Type I and Type II errors and Power
plt.fill_between(x, 0, pdf_null, where=(x > z_critical),
                 color='orange', alpha=0.5, label='Type I Error ()')
plt.fill_between(x, 0, pdf_alt, where=(x <= z_critical),
                 color='purple', alpha=0.5, label='Type II Error ()')
plt.fill_between(x, 0, pdf_alt, where=(x > z_critical),
                 color='green', alpha=0.5, label='Power (1-)')

# Critical value line
plt.axvline(z_critical, color='black', linestyle='--', linewidth=1.5, label=f'Critical Value (z = {z_critical:.2f})')

# Labels and legend
plt.title('Statistical Decisions: Type I and Type II Errors with Power', fontsize=14)
plt.xlabel('Test Statistic (z)', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.legend(fontsize=10)
plt.grid(alpha=0.3)
plt.tight_layout()

# Show plot
plt.show()
```

Two Overlapping Distributions:

Blue Curve: This represents the distribution under the null hypothesis (0), which is modeled as a normal distribution with a mean of 0.
Red Curve: This represents the distribution under the alternative hypothesis (1), with a mean of 2 (different from the null hypothesis).
Critical Region (Type I Error, ):

The critical region is the area beyond the critical value where 0 would be rejected. This region is shaded orange. It represents the probability of making a Type I Error, which occurs when we reject 0 when it is actually true.

Type II Error (): This error occurs when we fail to reject 0 when 1 is true. The area under the red curve before the critical value is shaded purple, indicating the Type II Error probability.

Power (1): Power refers to the ability of the test to correctly reject 0 when 1 is true. The area under the red curve beyond the critical value is shaded green, representing the power of the test.
Critical Value: The critical value is represented by the vertical dashed line. It separates the rejection region (for 0 ) from the region where we fail to reject 
0 .

Interpretation of the Graph:  

Type I Error (): If the test statistic falls in the orange shaded region, we reject 0 when it is actually true.

Type II Error (): If the test statistic falls in the purple shaded region, we fail to reject 0  when 1  is actually true.

Power (1): The green shaded region represents the probability of correctly rejecting 0  when 1  is true, i.e., the test's effectiveness.
