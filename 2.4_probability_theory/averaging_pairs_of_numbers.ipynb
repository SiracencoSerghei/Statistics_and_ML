# Averaging pairs of numbers
---

```python
import numpy as np
import matplotlib.pyplot as plt
```

variant 1

```python
# Parameters
min_val = 1   # Minimum value of uniform distribution
max_val = 100 # Maximum value of uniform distribution
num_samples = 100000  # Number of samples to generate

# Generate two sets of uniformly distributed integers
uniform1 = np.random.randint(min_val, max_val + 1, num_samples)
uniform2 = np.random.randint(min_val, max_val + 1, num_samples)

# Compute the average of the two sets
averages = (uniform1 + uniform2) / 2.0

# Plot histogram of averages
plt.figure(figsize=(10, 6))
plt.hist(averages, bins=range(min_val, max_val + 2), density=True, alpha=0.7, color='blue', edgecolor='black')
plt.title('Histogram of Averages of Two Uniformly Distributed Integers')
plt.xlabel('Average Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

variant 2

```python
import numpy as np
import matplotlib.pyplot as plt

# Sampling sizes
sampling_size = 2000
mean_sampling_size = 2000

# Arrays to store all averages and mean distributions
aver = np.zeros(100 * 100)
mean_dist = np.zeros(mean_sampling_size)

# Calculate all possible averages between 1 and 100
for i in range(1, 101):
    for ii in range(1, 101):
        aver[(i - 1) * 100 + (ii - 1)] = (i + ii) / 2

# Choose a random sample from the distribution of averages
aver_dist = np.random.choice(aver, sampling_size)

# Compute the mean of sampled averages multiple times
for i in range(mean_sampling_size):
    mean_dist[i] = np.mean(np.random.choice(aver_dist, sampling_size))

# Plotting
fig = plt.figure(figsize=(8, 5))
ax1 = plt.subplot2grid((2, 2), (0, 0), colspan=2)
ax2 = plt.subplot2grid((2, 2), (1, 0), colspan=1)
ax3 = plt.subplot2grid((2, 2), (1, 1), colspan=1)

# Plot all possible averages
ax1.plot(aver)
ax1.set_title('Population')

# Plot histogram of sampled averages
ax2.hist(aver_dist, bins=20)
ax2.set_title('Average Distribution')

# Plot histogram of means of sampled averages
ax3.hist(mean_dist, bins=40)
ax3.set_title('Mean Distribution')

plt.tight_layout()
plt.show()
```
