# SECTION: Probability theory
# cdf's and pdf's

```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
## example using log-normal distribution

# variable to evaluate the functions on
x = np.linspace(0,5,1001)

# note the function call pattern...
p1 = stats.lognorm.pdf(x,1)
c1 = stats.lognorm.cdf(x,1)

p2 = stats.lognorm.pdf(x,.1)
c2 = stats.lognorm.cdf(x,.1)
```

```python
# draw the pdfs
fig,ax = plt.subplots(2,1,figsize=(4,7))

ax[0].plot(x,p1/sum(p1)) # question: why divide by sum here?
ax[0].plot(x,p1/sum(p1), x,p2/sum(p2))
ax[0].set_ylabel('probability')
ax[0].set_title('pdf(x)')

# draw the cdfs
ax[1].plot(x,c1)
ax[1].plot(x,c1, x,c2)
ax[1].set_ylabel('probability')
ax[1].set_title('cdf(x)')
plt.show()
```

```python
## computing the cdf from the pdf

# compute the cdf
c1x = np.cumsum( p1*(x[1]-x[0]) )

plt.plot(x,c1)
plt.plot(x,c1x,'--')
plt.show()
```

```python
fig = plt.figure(constrained_layout= True,figsize=(10,10))
fig.suptitle('Plotting PDF and CDF for different distributions')
dist_list = [stats.lognorm,stats.norm,stats.uniform,stats.gamma]
subfigs = fig.subfigures(nrows=len(dist_list), ncols=1)
for row, (subfig,fn) in enumerate(zip(subfigs,dist_list)):
    subfig.suptitle(f'Plots for {fn.name}')

    # create 1x2 subplots per subfig
    axs = subfig.subplots(nrows=1, ncols=2)
    for j in [0.1,1,2,2.5]:
        pdf_dataset = fn.pdf(x,j)
        cdf_dataset = fn.cdf(x,j)
        axs[0].plot(x,pdf_dataset/sum(pdf_dataset),label=f'{j}')
        axs[1].plot(x,cdf_dataset,label=f'{j}')
        axs[0].set_title('pdf(x)')
        axs[1].set_title('cdf(x)')
        axs[0].legend(loc="upper right")
        axs[1].legend(loc="lower right")
```

```python
x = np.linspace(0,6,1001)



p_normal = stats.norm.pdf(x,3)

c_normal = stats.norm.cdf(x,3)



p_uniform = stats.uniform.pdf(x,3)

c_uniform = stats.uniform.cdf(x,3)



p_t =stats.t.pdf(x,3)

c_t = stats.t.cdf(x,3)



p_f =stats.f.pdf(x,3,3)

c_f = stats.f.cdf(x,3,3)



p_exp =stats.expon.pdf(x,3)

c_exp = stats.expon.cdf(x,3)





fig, ax = plt.subplots(5,2 ,figsize=(8,15))



#normal dis.

ax[0][0].plot(x,p_normal/sum(p_normal),color='r')

ax[0][0].set_title('Normal distribution (PDF)')

ax[0][0].set_ylabel('probability')

ax[0][1].plot(x,c_normal,color='r')

ax[0][1].set_title('Normal distribution (CDF)')



# uniform dis.



ax[1][0].plot(x,p_uniform/sum(p_uniform),'g')

ax[1][0].set_title('uniform distribution (PDF)')

ax[1][0].set_ylabel('probability')

ax[1][1].plot(x,c_uniform,'g')

ax[1][1].set_title('uniform distribution (CDF)')



# t-distribution

ax[2][0].plot(x,p_t/sum(p_t),'y')

ax[2][0].set_title('t-distribution (PDF)')

ax[2][0].set_ylabel('probability')

ax[2][1].plot(x,c_t,'y')

ax[2][1].set_title('t-distribution (CDF)')



ax[3][0].plot(x,p_f/sum(p_f))

ax[3][0].set_title('f-distribution (PDF)')

ax[3][0].set_ylabel('probability')



# f-distribution

ax[3][1].plot(x,c_f)

ax[3][1].set_title('f-distribution (CDF)')



#exponential distribution

ax[4][0].plot(x,p_exp/sum(p_exp),'m')

ax[4][0].set_title('exponential distribution (PDF)')

ax[4][0].set_ylabel('probability')



ax[4][1].plot(x,c_exp,'m')

ax[4][1].set_title('exponential distribution (CDF)')



plt.subplots_adjust(hspace=0.3)

plt.show()
```

```python
# define x axis
x = np.linspace(0,10,10001)

# pdf and cdf for normal distribution
p = stats.norm.pdf(x, 5)
d = stats.norm.cdf(x, 5)

# plotting
fig,ax = plt.subplots(2,1,figsize=(6,6))
ax[0].plot(x,p/sum(p))
ax[0].set_title("Probability Density Function of Gaussian distribution")
ax[0].set_ylabel("Probability")

ax[1].plot(x,d)
ax[1].set_title("Cumulative Density Function of Gaussian distribution")
ax[1].set_ylabel("Probability")

plt.subplots_adjust(hspace=0.4)

plt.show()

# check that sum of all probabilities equals to 1 (or close to 1)
print(f"\n Sum of all probabilities is {sum(p*(x[1]-x[0]))}")
```
