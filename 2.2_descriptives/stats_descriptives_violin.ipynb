# SECTION: Descriptive statistics
# Violin plots

```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
## create the data

n = 1000
thresh = 5 # threshold for cropping data

data = np.exp( np.random.randn(n) )
data[data>thresh] = thresh + np.random.randn(sum(data>thresh))*.1

plt.plot(data,'+')
```

```python
# show histogram
plt.hist(data,30)
plt.title('Histogram')
plt.show()

# show violin plot
plt.violinplot(data)
plt.title('Violin')
plt.show()
```

```python
# another option: swarm plot

import seaborn as sns
sns.swarmplot(data,orient='v', size=3)
```

```python
# another option: swarm plot
sns.swarmplot(data,orient='h', size=3)
```

# Code challenge: asymmetric violin plots

```python
import pandas as pd

n = 1000
thresh = 5

data = np.exp(np.random.randn(n))
data[data > thresh] = thresh + np.random.randn(sum(data > thresh)) * .1

data2 = np.random.randn(n) + 3

a = np.array(['Log-normal'] * n)
b = np.array(['Normal'] * n)

col3 = np.concatenate((a, b))
dataf = np.concatenate((data, data2))

c = np.array(['Dummy'] * n * 2)

# create 'tidy' dataframe
df = pd.DataFrame({'data1': dataf, 'Distribution': col3, 'dummy': c})

sns.violinplot(data=df, x='dummy', y='data1', hue='Distribution', orient='v', split=True, palette='Purples')

plt.show()
```

```python
N=100

data_1 = np.random.normal(loc=0, size=N)

data_2 = np.random.normal(loc=2, size=N)



data = pd.DataFrame({'data_1':data_1, 'data_2':data_2})

data = data.melt()

data['dummy'] = 0



sns.violinplot(data=data, x='dummy', y='value', split=True, hue='variable', palette='winter')
```
